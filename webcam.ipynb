```python
import cv2
import threading
import ipywidgets as widgets
from IPython.display import display, clear_output
import PIL.Image

# Create a button widget
capture_button = widgets.Button(description='Capture Image')
display(capture_button)

# Create an output widget to display the webcam feed
webcam_output = widgets.Output()
display(webcam_output)

# Create an output widget to display the captured image
captured_image_output = widgets.Output()
display(captured_image_output)

# Create a VideoCapture object to access the webcam
cap = cv2.VideoCapture(0)

# Variable to control the webcam display thread
webcam_thread_running = False

# Function to capture and display the webcam feed
def show_webcam():
    global webcam_thread_running
    try:
        while webcam_thread_running:
            ret, frame = cap.read()
            if not ret:
                break

            # Convert the OpenCV image to a format that can be displayed in IPython
            frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            pil_image = PIL.Image.fromarray(frame_rgb)

            # Display the webcam feed
            with webcam_output:
                clear_output(wait=True)
                display(pil_image)
                
    except KeyboardInterrupt:
        pass
    finally:
        cap.release()
        with webcam_output:
            clear_output(wait=True)

# Function to capture an image and display it
def capture_image(_):
    global webcam_thread_running
    webcam_thread_running = False  # Stop the webcam thread
    
    ret, frame = cap.read()
    if ret:
        # Convert the OpenCV image to a format that can be displayed in IPython
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        pil_image = PIL.Image.fromarray(frame_rgb)

        # Display the captured image
        with captured_image_output:
            clear_output(wait=True)
            display(pil_image)

        # Save the captured image
#         image_filename = "/photo.jpg"
#         cv2.imwrite(image_filename, frame)
#         print(f"Image saved as {image_filename}")


# Link the button click events to the appropriate functions
capture_button.on_click(capture_image)

# Start showing the webcam feed in a separate thread
webcam_thread_running = True
webcam_thread = threading.Thread(target=show_webcam)
webcam_thread.start()
```
